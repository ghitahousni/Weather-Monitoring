Then, we will be using these URLs to download the weather data from the corresponding stations. To do that, we first need to define our download function.

```python
#We first define our download function

def download(url, folder="", filename=None):
    """
    Download weather data from Met Office. 
    """

    #We provide a name to our files
    if filename is None: 
        #Use default file name  
        filename = os.path.basename(urllib.parse.urlparse(url).path)
            
    #We join the specified folder to the determined filename into a filepath 
    filepath = os.path.join(folder, filename) 
    
    #If the filepath already exists, we print the below message
    if os.path.isfile(filepath): 
        print(f'File {filepath} already exists')
        return filepath
    
    #If the filepath needs to be downloaded, we print the downloaded message
    components = urllib.parse.urlparse(url)
    print(f"Downloading '{os.path.basename(components.path)}' from {components.netloc}")

    #Finally, we show how long the file took to download
    t0 = time.time()
    try:
        urllib.request.urlretrieve(url, filepath)
    except KeyboardInterrupt: 
        if os.path.exists(filepath): 
            #Try remove downloaded file 
            os.remove(filepath)
        raise

    dt = time.time() - t0 
    
    #We print our result
    print(f"Download complete ({dt: .2f}s)")
    return filepath
```

##### Code explanation: 

We first define our download function that is designed to download the weather data from the specified URLs and save it to our folder. Our download function intend to download weather data from the the specified Met Office URLs and save it to our folder. It will take 3 parameters: 

- url : the URL from which to download data, that we previously generated
- folder : the folder where the downloaded file will be saved
- filename : the specific name we will be giving to our files

If the filename is not provided, we ask the function to extract it from the URL. After that, we ask the function to check if the file already exists in the specified location (os.path.isfile(filepath)). If it does, we ask the function to print a message, and returns the filepath without downloading the file again. We also ask the function to record the current time (t0) and to attempt to retrieve the file from the specified URL using urllib.request.urlretrieve(url, filepath). If a keyboard interrupt occurs during the download, we ask the function to remove the partially downloaded file before raising the interrupt again.
Finally, the function calculates the time taken for the download (dt), prints a completion message with the time duration and returns the filepath of the downloaded file.

However, in order to use them for our code, we need to convert them into csv file through a load function that we will be defining below. 

```python
#We now define our load function 

def load(filename, stationfile, weatherfile): 
    """
    Returns data from weather stations
    """
    
    #First, we open and read the file
    with open(filename, 'r') as first_file: 
        stripped = (line.strip() for line in first_file) #Then, we turn it into line strips
        lines = (line.split() for line in stripped if line.strip()) #Finally we turn each line into a list of values
        
        #Then, we create an intermediate file (stationfile)
        with open(stationfile, 'w', newline = '') as second_file: 
            writer = csv.writer(second_file)
            if lines: 
                writer.writerow(['yyyy', 'month', 'tmax(degC)', 'tmin(degC)', 'af(days)', 'rain(mm)', 'sun(hours)'])
                writer.writerows(lines)
                
    #We then re-open the intermediate file (stationfile)
    #And create our final required file (weatherfile) based on the above
    with open(stationfile, 'r', newline = '') as first_file: 
        with open(weatherfile, 'w', newline = '') as second_file: 
            writer = csv.writer(second_file)
            for row in csv.reader(first_file):
                if row: 
                    writer.writerow(row)
    
    #Finally, we delete the intermediate file (stationfile)
    os.remove(stationfile)
```

##### Code explanation:

Our load function consist in loading the required data from our previous downloaded files. It is designed to process our downloaded text files, specified by the parameter "filename", reformat its content into an intermediate file specified by the parameter "stationfile", and save it to a new file specified by the parameter "weatherfile".

To do that, we first ask the function to open and read the file specified by "filename" parameter. We then ask it to convert it into line strips, removing whitespaces and empty lines. Finally, we ask it to turn each line into a list of values. Then, we ask the function to create an intermediate file (stationfile) where the stripped and formatted data is written and relabelled into columns. After that, we ask the function to reopen the intermediate file ("stationfile") for reading. The function will format each non-empty row and writte it into a new final version file ("weatherfile"). Finally, once the data is reformatted and saved into our final version file ("weatherfile"), we ask the function to delete our intermedial file ("stationfile") as it is unnecessary. This explains why we only have our text file and the csv weatherfile loaded into our folder.

#### C) Select a monitoring time period when your data from all your selected weather stations have been consistently recorded. 

As seen in our previously generated dataframes, the monitoring time periods differ from each station. Therefore, the time period that we are choosing is from 1998 to 2023, as it spans over two decades and include the 21st century. This will allow us to examine trends, variations and shifts in weather patterns over a long-term where the data seems reliable and consistent across the 37 weather stations. This way, we will be able to provide a comprehensive view of weather patterns at a national level. 

Therefore, we first want to find our specified range in terms of rows (slicing range) for all our 37 dataframes. 

```python
#We first initialize a list to store our row ranges 
range_list = []

#We then iterate over our cleaned DataFrames 
for i, df in enumerate(clean_dataframe_list):
    #We convert the 'yyyy' column to numeric, handling errors by coercing NaN
    df['yyyy'] = pd.to_numeric(df['yyyy'], errors='coerce') 
    
    #We define our chosen time period
    start_year = 1998
    end_year = 2023

    #We filter the rows based on the selected time period
    selected_rows = df[(df['yyyy'] >= start_year) & (df['yyyy'] <= end_year)]

    #We first check whether there are any rows in the selected time period
    if not selected_rows.empty: 
        #We then get the minimum and maximum indices to represent the full range of rows
        range_start = selected_rows.index.min()
        range_end = selected_rows.index.max()
        row_range = (range_start,range_end)
        #We append the tuple representing the range row to the range_list
        range_list.append(row_range) 
        
        #We then print the row_range for the selected time period 
        print(f"In dataframe{i}, years {start_year} to {end_year} are represented by rows: {row_range}")

    else: 
        ##If no rows are found, we print the below information
        print(f"No rows found with the year {start_year} to {end_year}:")
```

Now that we sliced our dataframes according to our year range (1998-2023), we need to slice each individual parameter from each dataframe in order to find the tmax(degC), tmin(degC), af(days), rain(mm) and sun(hours) averages per station.

We first want to calculate the mean of tmax(degC) for each station on our sliced dataframes

```python
#We initialize a list where to store our tmax(degC) averages and station names
tmax_average_list = []

#We iterate over our sliced dataframes and their corresponding indices
for i, sliced_dataframe in enumerate(sliced_dataframe_list):
    #We then extract 'tmax(degC)' column/parameter from our dataframe
    tmax_dataframe = sliced_dataframe.loc[:, 'tmax(degC)']
    #We convert the 'tmax(degC)' column into numeric to handle errors
    tmax_dataframe = pd.to_numeric(tmax_dataframe, errors='coerce')  
    #We then drop NaN values
    tmax_dataframe = tmax_dataframe.dropna()  
    #Finally we calculate the mean of the tmax(degC) column for each station
    tmax_average = statistics.mean(tmax_dataframe) 
    #We extract our station names from our filename
    station_name = filename_list[i].split('data.txt')[0] 
    #And append the station name to its corresponding average tmax(degC)
    tmax_average_list.append((f"Station {station_name}", tmax_average)) 

#Finally, we iterate over our tmax_average_list and print the results
for station, tmax_average in tmax_average_list:
    print(f"Average maximum temperature in {station}: {tmax_average}")
```

#### A) Calculate for each station the average temperature change to date as a percentage increase or decrease.

In order to calculate each station average temperature change, we need to use our previous results from Question 1. Indeed, we first want to find the average temperature change. To do that, we will be using our tmax_average_list and tmin_average list. These lists contain average tmax(degC) and tmin(degC) values over a period of 25 years (1998-2023). Once done, we would need to calculate the average percentage change. To do that, we will use the below formula: 

                             percentage change = ((Tmax - Tmin) / tmax)) * 100    

```python
#We will be using tmax_average_list and tmin_average_list defined in Q1

#We first initialize an empty list to store our percentage changes
change_percentages_list = []

tmax_avg = [tmax[1] for tmax in tmax_average_list]
tmin_avg = [tmin[1] for tmin in tmin_average_list]

#Then, we iterate through our lists and calculate the average temperature change as a percentage
for i, (tmax, tmin) in enumerate(zip(tmax_avg, tmin_avg)):
    #We then name our stations 
    station_name = filename_list[i].split('data.txt')[0]
    
    #We calculate temperature change and percentage change
    temperature_change = tmax - tmin 
    
    #We check if tmax is not 0 to avoid division by zero
    if tmax != 0:
       
        #As per our percentage change formula
        percentage_change = (temperature_change / tmax) * 100
        
        #And append the result for each station
        change_percentages_list.append((f"Station {station_name}", percentage_change))
        
        #Finally, we print each station name and its corresponding average temperature percentage change
        print(f"Percentage change for Station {station_name}: {percentage_change: }%") 
    else:
        #We print an error message if Tmax is 0
        print(f"Error: Tmax is 0 for station {station_name}.")
```

#### Code explanation: 

The above code aims to find the average temperature change as a percentage for each station. We now want to find the mean Tmax and Tmin values of all the stations aggregated by year over 1998 to 2023 to understand the tmax(degC) and tmin(degC) mean values progression over the years (1998-2023).

To do that, we first have to find all tmax(degC) averages per year. 

```python
#We first initialize an empty list to store all our tmax(degC) averages
average_tmax_by_year_list = []

#We then iterate through our clean_data_frame list 
for i in range(len(clean_dataframe_list)): 
    #We extract our dataframe and corresponding range
    df = clean_dataframe_list[i]
    start, end = range_list[i]
    #We then slice our dataframe based on our corresponding range
    sliced_dataframe = df.iloc[start:end +1]
    
    #We then keep only our 'yyyy' and 'tmax(degC)' columns 
    sliced_dataframe = sliced_dataframe[['yyyy', 'tmax(degC)']]
    #We convert our tmin(degC) values to float so we can work with them
    sliced_dataframe['tmax(degC)'] = sliced_dataframe['tmax(degC)'].astype(float)
    
                                     
    #We group our 'yyyy' and calculate the mean of tmin(degC) for each year
    average_tmax_by_year = sliced_dataframe.groupby('yyyy')['tmax(degC)'].mean()
    #We append the resulting series to our defined list
    average_tmax_by_year_list.append(average_tmax_by_year)
    
    #Finally, we display the result for each year for each station
    print(f"Average Tmax(degC) DataFrame for {i+1}: \n {average_tmax_by_year} \n")
```

#### Code explanation: 

Our above code calculates the average tmac(degC) for each year across multiple stations. It initializes an empty list, average_tmax_by_year_list, to store the average tmax values for each station. It then iterates through our cleaned_dataframes list, extracting each dataframe along with its corresponding range. Our dataframe is sliced based on the specified range, retaining only the 'yyyy' and 'tmax(degC)' columns. Then, our 'tmax(degC)' values are converted to float numbers for numerical analysis. We then group the sliced dataframe by year and calculates the mean tmax for each year. The resulting series is appended to the average_tmax_by_year_list. Finally, our code prints the average tmax values for each year for each station.

We now do the same for our tmin(degC) averages per year

```python
#We first initialize an empty list to store all our tmin(degC) averages
average_tmin_by_year_list = []

#We then iterate through our clean_data_frame list 
for i in range(len(clean_dataframe_list)): 
    #We extract our dataframe and corresponding range
    df = clean_dataframe_list[i]
    start, end = range_list[i]
    #We then slice our dataframe based on our corresponding range
    sliced_dataframe = df.iloc[start:end +1]
    
    #We then keep only our 'yyyy' and 'tmin(degC)' columns 
    sliced_dataframe = sliced_dataframe[['yyyy', 'tmin(degC)']]
    #We convert our tmin(degC) values to float so we can work with them
    sliced_dataframe['tmin(degC)'] = sliced_dataframe['tmin(degC)'].astype(float)
                     
    #Finally, we group our 'yyyy' and calculate the mean of tmin(degC) for each year
    average_tmin_by_year = sliced_dataframe.groupby('yyyy')['tmin(degC)'].mean()
    #We append the resulting series to our defined list
    average_tmin_by_year_list.append(average_tmin_by_year)
    
    #Finally, we display the result for each year for each station
    print(f"Average Tmin(degC) DataFrame for {i+1}: \n {average_tmin_by_year} \n")
```

Now we  can find the mean Tmax and Tmin values of all the stations aggregated by year over 1998 to 2023.

```python
#We first initialize an empty list to store the mean values for each year
meanmax_values_by_year = []

#We then loop through each year from 1998 to 2023
for year in range(1998, 2024):
    #We initialize an empty list to store the values for the current year
    values_for_year = []

    #Then we loop through each series/stations in the list
    for series in average_tmax_by_year_list:  
        ##We then extract all values for the current year fro each station and append them to the list
        values = series.loc[series.index == year].values
        values_for_year.extend(values)

    #We then calculate the mean of all values for the current year 
    mean_for_year = pd.Series(values_for_year).mean()
    meanmax_values_by_year.append(mean_for_year)

#Finally, we create a dataframe to display the results
average_tmax_values_per_year = pd.DataFrame({'Year': range(1998, 2024), 'Mean_Value': meanmax_values_by_year})

#And print it
print(average_tmax_values_per_year)

#Then we try to find the global average Tmax accross all years
global_tmax_average = np.mean(meanmax_values_by_year) 

print(f"The average of Tmax from 1998 to 2023 is : {global_tmax_average: .2f}degC")
```

#### Code explanation: 

This above code allows us to get a summary of temperature trends over our specified time period. Indeed, it calculates the mean maximum temperatures (tmax) for each year from 1998 to 2023 across multiple stations.
